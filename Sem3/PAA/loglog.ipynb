```python
import math
import matplotlib.pyplot as plt

class AlgorithmAnalys:
    def __init__(self) -> None:
        self.results = []
        self.indexes = []
        self.Ni = 0
    
    def add_result(self, index, result):
        self.results.append(result)
        # self.Ni += 1
        # self.indexes.append(self.Ni)
        self.indexes.append(index)
        
    def plot(self, color = "red"):        
        plt.plot(self.indexes, self.results, color = color)

        plt.title("Algorithm")
        plt.xlabel("k")
        plt.ylabel("Nsteps")

        # Set x-axis ticks to integers with step 1
        # plt.xticks(range(1, self.Ni + 1, 1))
        # plt.yticks(range(1, self.Ni * 3, 1))

        plt.show()
```

```python
def Loglog(k, n, analys):
    Nsteps = 0
    
    while (k < n):
        k = k * k
        Nsteps += 1
        analys.add_result(k , Nsteps)

    analys.plot()

    return Nsteps
```

```python
analys = AlgorithmAnalys()

Loglog(2, 10000, analys)
```
